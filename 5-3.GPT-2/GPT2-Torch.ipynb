```
'''
  code by Minho Ryu @bzantium
  
'''
import numpy as np
import torch
from pytorch_pretrained_bert import GPT2Tokenizer, GPT2LMHeadModel
from torch.nn import functional as F

tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')

sequence = tokenizer.encode("Fake news or junk news or pseudo-news is a type of yellow journalism or propaganda that consists of " + 
                            "deliberate disinformation or hoaxes spread via traditional print and broadcast news media or online social media.")
_input, past = torch.tensor([sequence]), None

output = []
out_maxlen = 100

for _ in range(out_maxlen):
    logits, past = model(_input, past=past)
    _input = torch.argmax(F.softmax(logits[:, -1], dim=-1), dim=-1, keepdim=True)
    # _input = torch.multinomial(F.softmax(logits[:, -1], dim=-1), 1) # for the diverse text generation
    if _input.item() == 198: # Remove \n
        continue
    output.append(_input.item()) 
    if tokenizer.decode([_input.item()])[-1] == '.': # we stop generating when dot(.) appears
        break
    

print('Context:', tokenizer.decode(sequence))
print('Generated:', tokenizer.decode(output))
```
